```python
## imports
import numpy as np
import pandas as pd
from sklearn import metrics##used to show the accuracy
from sklearn.metrics import confusion_matrix #used to describe performance of classifier
from sklearn.metrics import classification_report##info abouve accuracy, precision and so on...
from sklearn.model_selection import train_test_split## allows us to split the given data 
from sklearn.feature_extraction.text import CountVectorizer##converts text documents to matrix
```

```python
## our data imported
info = pd.read_excel(r'C:\Users\ASUS\Desktop\python\azeri_news.xlsx',sep='\t',names=['Category','Title','news_article'])
```

```python
info.head()## first rows from our data 
```

```python
df_x=info["news_article"]+info["Title"]
df_y=info["Category"]
```

```python
## Now performing the same with TFID vectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
ve = TfidfVectorizer()
```

```python
ve.vocabulary_
## mapping of terms to idices of features
```

```python
x_train, x_test, y_train, y_test = train_test_split(df_x, df_y, test_size=0.2, random_state=4)
```

```python
x_train.head()
```

```python
x_traincv = ve.fit_transform(x_train)
```

```python
##performing SVM now
from sklearn.svm import SVC
```

```python
## with kernel = 'RBF' and gamma = 'scale' the accuracy was 82%
## using cache size to speed up the run process as it takes veeery long
clf = SVC(kernel='linear',gamma= 'auto', C=1, cache_size=7000, random_state = 1)
```

```python
clf.fit(x_traincv,y_train)
```

```python
cv_t = ve.transform(x_test)
SVM_predicted =clf.predict(cv_t)
```

```python
SVM_predicted
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, SVM_predicted))
```

```python
print(classification_report(y_test,SVM_predicted))
```

```python
confusion_matrix(y_test, SVM_predicted)
```

```python
##importing neural network
##without any hyperparameters the accuracy was only 76 
from sklearn.neural_network import MLPClassifier
mc = MLPClassifier(hidden_layer_sizes=(10,10,10),activation='relu', solver='adam', random_state = 1, max_iter=200)
```

```python
mc.fit(x_traincv, y_train)
```

```python
cv_t = ve.transform(x_test)
NN_prediction1 = mc.predict(cv_t)
```

```python
NN_prediction1
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, NN_prediction1))
```

```python
print(classification_report(y_test,NN_prediction1))
```

```python
confusion_matrix(y_test, NN_prediction1)
```
