```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
##importing neural network
from sklearn.neural_network import MLPClassifier
##importing Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier
##importing Support Vector Machine
from sklearn.svm import SVC
## metrics to print out the matrix, F1, recall, precision and accuracy after training the data
from sklearn.metrics import classification_report
from sklearn import metrics
##imports for extra task (shuffler and model selector fo)
from sklearn.utils import shuffle as shuffler
from sklearn import model_selection
```

```python
##loading the data by getting rid of ; and # using the sep function of pandas library
train_data = pd.read_csv(r'C:\Users\ASUS\Desktop\data\Train.txt', header=None, sep=';|#', engine='python')
cross_valid_data = pd.read_csv(r'C:\Users\ASUS\Desktop\data\Cross_Validation.txt',header=None, sep=';|#', engine='python')
test_data = pd.read_csv(r'C:\Users\ASUS\Desktop\data\Test.txt',header=None, sep=';|#', engine='python')
```

```python
## Train data visualization
train_X = train_data.iloc[:,1:]
train_Y = train_data.iloc[:,0]
print(train_X,train_Y)
```

```python
## Cross Validation data visualization
crosv_X=cross_valid_data.iloc[:,1:]
crosv_Y=cross_valid_data.iloc[:,0]
print(crosv_X,crosv_Y)
```

```python
##Test data visualization
test_X=test_data.iloc[:,1:]
test_Y=test_data.iloc[:,0]
print(test_X,test_Y)
```

```python
## Training the model by Neural Network
mc = MLPClassifier(hidden_layer_sizes=(80,80,80),activation='relu', solver='adam',max_iter=500)
mc.fit(train_X, train_Y)

NN_prediction1 = mc.predict(crosv_X)
```

```python
print("Accuracy:",metrics.accuracy_score(crosv_Y,NN_prediction1))
```

```python
print(classification_report(crosv_Y,NN_prediction1))
```

```python
NN_prediction2=mc.predict(test_X)
```

```python
print("Accuracy:",metrics.accuracy_score(test_Y,NN_prediction2))
```

```python
print(classification_report(test_Y,NN_prediction2))
```

```python
## Training the model by SVM
clsf = SVC(kernel='rbf',gamma= 'scale', C=1)## tried several parameters kernel and gamma (gamma scale produced the highest accuracy value)
clsf.fit(train_X, train_Y)
```

```python
SVM_predicted =clsf.predict(test_X)
```

```python
print("Accuracy:",metrics.accuracy_score(test_Y, SVM_predicted))
```

```python
print(classification_report(test_Y,SVM_predicted))
```

```python
SVM_predicted2=clsf.predict(crosv_X)
```

```python
print("Accuracy:",metrics.accuracy_score(crosv_Y,SVM_predicted2))
```

```python
print(classification_report(crosv_Y,SVM_predicted2))
```

```python
## Training the model by Decision Tree Classifier
dtree = DecisionTreeClassifier(criterion="entropy", splitter = "best")
dtree.fit(train_X, train_Y)
```

```python
DTC_prediction = dtree.predict(crosv_X)
```

```python
print("Accuracy:",metrics.accuracy_score(crosv_Y,DTC_prediction))
```

```python
print(classification_report(crosv_Y,DTC_prediction))
```

```python
DTC_prediction2=dtree.predict(test_X)
```

```python
print("Accuracy:",metrics.accuracy_score(test_Y,DTC_prediction2))
```

```python
print(classification_report(test_Y,DTC_prediction2))
```

```python
## EXTRA TASK
result = pd.concat([train_data,cross_valid_data,test_data])
print(result)
result = shuffler(result)
x = result.iloc[:,1:]
y = result.iloc[:,0]
```

```python
kf = model_selection.KFold(n_splits=10, random_state=None)
cv_res = model_selection.cross_val_score(MLPClassifier(), x, y, cv=kf)
print("Accuracy: " , (cv_res.mean()))
```

```python
cv_res1 = model_selection.cross_val_score(DecisionTreeClassifier(), x, y, cv=kf)
print("Accuracy: " , (cv_res1.mean()))
```

```python
cv_res2 = model_selection.cross_val_score(SVC(), x, y, cv=kf)
print("Accuracy: " , (cv_res2.mean()))
```
